```python
import os

import pandas as pd
import pyspark.pandas as ps
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv(os.path.join("..", "data", "otodom_wroclaw.csv"))
```

```python
df.shape
```

```python
df2 = pd.concat([df for _ in range(1)])
```

```python
df2.shape
```

```python
data = df2.copy()

# Drop columns with too many missing values or irrelevant information
data = data.drop(columns=['address_city', 'address_province', 'address_street', 'advertiser_type'])

# Handle missing values and encoding categorical variables
numerical_features = ['build_year', 'price', 'rent', 'rooms_num', 'surface']
categorical_features = ['address_district', 'address_estate', 'building_material', 'building_ownership', 'building_type', 'heating', 'market', 'media', 'outdoor', 'parking_space', 'securities', 'windows']

numerical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features)
    ])

# Apply transformations
data_preprocessed = preprocessor.fit_transform(data)

# Convert the preprocessed data into a DataFrame for easier manipulation
data_preprocessed = pd.DataFrame(data_preprocessed.toarray())
data_preprocessed["id"] = range(len(data_preprocessed))

# Display the first few rows of the preprocessed data
data_preprocessed.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Compute the cosine similarity matrix
similarity_matrix = cosine_similarity(data_preprocessed)

# Function to get the most similar offers
def get_similar_offers(chosen_offer_index, top_n=5):
    similarity_scores = list(enumerate(similarity_matrix[chosen_offer_index]))
    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)
    most_similar_indices = [index for index, score in similarity_scores[1:top_n+1]]
    return data.iloc[most_similar_indices]

# Example: Get the top 5 offers similar to the first offer
similar_offers = get_similar_offers(chosen_offer_index=0, top_n=5)
similar_offers
```

```python
data_preprocessed_spark = ps.from_pandas(data_preprocessed.iloc[:9999, :]).to_spark()
```

```python
from functools import reduce
from operator import add

from pyspark.sql import SparkSession, functions as F, types as T


def mynorm(*cols):
    return F.sqrt(reduce(add, [F.pow(c, 2) for c in cols]))    
    
def mydot(a_cols, b_cols):
    return reduce(add, [a * b for a, b in zip(a_cols, b_cols)])


def cos_sim_ps(a_cols, b_cols):
    return mydot(a_cols, b_cols) / (mynorm(*a_cols) * mynorm(*b_cols))


similarity_matrix = (
    data_preprocessed_spark.alias('a')
    .crossJoin(data_preprocessed_spark.alias('b'))
    .withColumn(
        'cs',
        cos_sim_ps(
            [F.col(f'a.{c}') for c in data_preprocessed_spark.columns],
            [F.col(f'b.{c}') for c in data_preprocessed_spark.columns],
        )
    )
    .groupby('a.id')
    .pivot('b.id')
    .sum('cs')
)
```
